```python
#plane

from scipy import misc
 
IMAGE = 'plane.jpg'
 
T = [192, 192, 192]
S = [135, 206, 250]

def dist(x, y):
    return sum((x - y) ** 2)

def flip(i, j, A, B):
    if (i, j) in A:
        A.remove((i, j))
        B.add((i, j))
    else:
        B.remove((i, j))
        A.add((i, j))

def cut(image):
    dim_x, dim_y, colors = image.shape
    
    currFirst = T
    currSecond = S
    lastFirst = [0, 0, 0]
    lastSecond = [0, 0, 0]
    
    A = set()
    B = set()
    
    while ((lastFirst, lastSecond) != (currFirst, currSecond)):
        A = set()
        B = {(i, j) for i in range(dim_x) for j in range(dim_y)}
        for i in range(dim_x):
            for j in range(dim_y):
                if (dist(image[i, j], currFirst) < dist(image[i, j], currSecond)):
                    flip(i, j, A, B)
        sum1 = [0, 0, 0]
        for elem in A:
            color = image[elem[0], elem[1]]
            sum1[0] += color[0]
            sum1[1] += color[1]
            sum1[2] += color[2]
        sum2 = [0, 0, 0]
        for elem in B:
            color = image[elem[0], elem[1]]
            sum2[0] += color[0]
            sum2[1] += color[1]
            sum2[2] += color[2]
            
        sum1[0] = sum1[0] / len(A)
        sum1[1] = sum1[1] / len(A)
        sum1[2] = sum1[2] / len(A)
        sum2[0] = sum2[0] / len(B)
        sum2[1] = sum2[1] / len(B)
        sum2[2] = sum2[2] / len(B)
        
        lastFirst = currFirst
        lastSecond = currSecond
        currFirst = sum1
        currSecond = sum2
     
    return A, B
 
image = misc.imread(IMAGE)
dim_x, dim_y, colors = image.shape    
print("Loaded image of shape {x}, {y}".format(x=dim_x, y=dim_y))
 
A, B = cut(image)
 
for i in range(dim_x):
    for j in range(dim_y):
        if (i, j) in A:                
            image[i, j, 0] = min(255, image[i, j, 0] + 100)
            image[i, j, 1] = image[i, j, 1] / 3
            image[i, j, 2] = image[i, j, 2] / 3
        elif (i,j) in B:
            image[i, j, 0] = image[i, j, 0] / 3
            image[i, j, 1] = image[i, j, 1] / 3
            image[i, j, 2] = min(255, image[i, j, 2] + 100)
misc.imsave('out_plane.png', image)
```

```python
#cat

from scipy import misc
 
IMAGE = 'cat.jpg'
 
S = [192, 192, 192]
T = [255, 255, 255]
P = [255, 204, 179]
V = [51, 26, 0]

def dist(x, y):
    return sum((x - y) ** 2)

def flip(i, j, A, B):
    if (i, j) in A:
        A.remove((i, j))
        B.add((i, j))
    else:
        B.remove((i, j))
        A.add((i, j))

def cut(image):
    dim_x, dim_y, colors = image.shape
    
    currFirst = T
    currSecond = S
    currThird = P
    currFourth = V
    lastFirst = [0, 0, 0]
    lastSecond = [0, 0, 0]
    lastThird = [0, 0, 0]
    lastFourth = [0, 0, 0]
    
    A = set()
    B = set()
    C = set()
    D = set()
    
    while ((lastFirst, lastSecond, lastThird, lastFourth) != (currFirst, currSecond, currThird, currFourth)):
        A = set()
        B = set()
        C = set()
        D = {(i, j) for i in range(dim_x) for j in range(dim_y)}
        for i in range(dim_x):
            for j in range(dim_y):
                if (min(dist(image[i, j], currFirst), dist(image[i, j], currSecond),
                        dist(image[i, j], currThird), dist(image[i, j], currFourth)) == dist(image[i, j], currFirst)):
                    flip(i, j, A, D)
                elif (min(dist(image[i, j], currFirst), dist(image[i, j], currSecond),
                        dist(image[i, j], currThird), dist(image[i, j], currFourth)) == dist(image[i, j], currSecond)):
                    flip(i, j, B, D)
                elif (min(dist(image[i, j], currFirst), dist(image[i, j], currSecond),
                        dist(image[i, j], currThird), dist(image[i, j], currFourth)) == dist(image[i, j], currThird)):
                    flip(i, j, C, D)
        sum1 = [0, 0, 0]
        for elem in A:
            color = image[elem[0], elem[1]]
            sum1[0] += color[0]
            sum1[1] += color[1]
            sum1[2] += color[2]
        sum2 = [0, 0, 0]
        for elem in B:
            color = image[elem[0], elem[1]]
            sum2[0] += color[0]
            sum2[1] += color[1]
            sum2[2] += color[2]
        sum3 = [0, 0, 0]
        for elem in C:
            color = image[elem[0], elem[1]]
            sum3[0] += color[0]
            sum3[1] += color[1]
            sum3[2] += color[2]
        sum4 = [0, 0, 0]
        for elem in C:
            color = image[elem[0], elem[1]]
            sum4[0] += color[0]
            sum4[1] += color[1]
            sum4[2] += color[2]
            
        sum1[0] = sum1[0] / len(A)
        sum1[1] = sum1[1] / len(A)
        sum1[2] = sum1[2] / len(A)
        sum2[0] = sum2[0] / len(B)
        sum2[1] = sum2[1] / len(B)
        sum2[2] = sum2[2] / len(B)
        sum3[0] = sum3[0] / len(C)
        sum3[1] = sum3[1] / len(C)
        sum3[2] = sum3[2] / len(C)
        sum4[0] = sum4[0] / len(D)
        sum4[1] = sum4[1] / len(D)
        sum4[2] = sum4[2] / len(D)
        
        lastFirst = currFirst
        lastSecond = currSecond
        lastThird = currThird
        lastFourth = currFourth
        currFirst = sum1
        currSecond = sum2
        currThird = sum3
        currFourth = sum4
     
    return A, B, C, D
 
image = misc.imread(IMAGE)
dim_x, dim_y, colors = image.shape    
print("Loaded image of shape {x}, {y}".format(x=dim_x, y=dim_y))
 
A, B, C, D = cut(image)
 
for i in range(dim_x):
    for j in range(dim_y):
        if (i, j) in A:                
            image[i, j, 0] = 255
            image[i, j, 1] = 255
            image[i, j, 2] = 255
        elif (i,j) in B:
            image[i, j, 0] = 192
            image[i, j, 1] = 192
            image[i, j, 2] = 192
        elif (i,j) in C:
            image[i, j, 0] = 255
            image[i, j, 1] = 204
            image[i, j, 2] = 179
        else:
            image[i, j, 0] = 51
            image[i, j, 1] = 26
            image[i, j, 2] = 0
misc.imsave('out_cat.png', image)
```
